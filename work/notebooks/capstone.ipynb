# Abstract

This project aims to identify content pages that should be refreshed using search performance data from the FlyRank Machine Learning Internship dataset. A rule-based baseline and a Random Forest model were developed and compared using the same train-test split. The baseline achieved an accuracy of approximately 0.70, while the Random Forest model achieved approximately 0.689 with higher recall. The findings suggest that simple rule-based approaches remain competitive while machine learning provides additional insights for identifying content refresh opportunities. The results are intended to support content prioritization and decision-making rather than prove causal ranking improvements.

# Introduction

Search rankings and user engagement change over time as content becomes outdated or competitors publish better information.

The objective of this project is to identify pages that are suitable candidates for content refresh using historical search performance metrics.

The project supports content managers by ranking pages that are likely to benefit from updates while avoiding future information leakage.

# Data

The project uses the **FlyRank Machine Learning Internship dataset** (`content_refresh_anonymized.csv`).

Dataset summary:

- Rows: 30,000
- Columns: 44
- One row represents one content page.
- Personally identifiable information and client names are excluded.
- Historical search performance metrics are used.
- Future labels are not used as input features.

# Methodology

This project follows the **Refresh / Content Opportunity Scoring** lane.

The workflow consists of:

1. Exploratory Data Analysis
2. Rule-based baseline scoring
3. Random Forest classification model
4. Model evaluation using an 80/20 train-test split
5. Comparison between the baseline and machine learning model

Features used include:

- Search Volume
- Competition
- CPC
- Word Count
- Character Count
- CTR
- Average Position
- Content Age
- Sessions
- Clicks
- Users

The model predicts whether a page should be considered for refresh.

# Results

The Week 4 baseline and the Random Forest model were evaluated using the same train-test split.

| Method | Accuracy |
|---------|----------|
| Baseline | 0.700 |
| Random Forest | 0.689 |

Additional metrics for the Random Forest model:

- Precision: 0.700
- Recall: 0.752
- F1 Score: 0.725

The baseline achieved slightly higher accuracy, while the Random Forest model achieved higher recall, identifying more potential refresh opportunities.

```python
import matplotlib.pyplot as plt

methods = ["Baseline", "Random Forest"]
accuracy = [0.70, 0.689]

plt.figure(figsize=(5,4))
plt.bar(methods, accuracy)
plt.title("Model Comparison")
plt.ylabel("Accuracy")
plt.ylim(0.6, 0.8)
plt.show()
```

# Limitations

This project uses historical search performance metrics and does not attempt to prove causal ranking improvements.

Some features contain missing values that were filled before training.

The baseline rule achieved slightly higher accuracy than the Random Forest model, indicating that simple decision rules remain competitive for this dataset.

Future work could include additional feature engineering, hyperparameter tuning, and evaluation on new data.

# Ranked Recommendations

Based on the analysis, the following actions are recommended:

1. Refresh pages with low CTR and high search volume.
2. Update outdated content that has declined in ranking.
3. Improve titles and meta descriptions for low CTR pages.
4. Monitor pages with improving trends before making changes.
5. Review pages with consistently low engagement for possible rewriting or consolidation.

These recommendations are intended to support editorial decision-making.
