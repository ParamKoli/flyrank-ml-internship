## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

## Research Question

### Decision
Which content pages should be reviewed first for refresh?

### Action
Content managers review the highest-ranked pages and decide whether to update, rewrite, expand, or monitor them.

### Cost of a Wrong Recommendation
A false positive wastes editorial time by recommending pages that do not need updating.

A false negative misses pages that actually require attention, leading to lost traffic and missed opportunities.

### Why Machine Learning
Machine learning can combine multiple search and engagement signals to produce better rankings than simple rule-based methods.
