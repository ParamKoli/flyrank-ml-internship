## 1. My rule and its reason codes

### Rule

The baseline rule identifies pages that are good candidates for content refresh.

A page receives a higher score if it has:
- Poor average search position (greater than 20)
- Low CTR (less than 2%)
- High search volume (greater than 100)
- Older content (more than 180 days)

These pages are likely to benefit from optimization because they already have search demand but are not performing well.

### Reason Codes

REFRESH_LOW_CTR

REFRESH_LOW_POSITION

REFRESH_HIGH_VOLUME

REFRESH_OLD_CONTENT

```python
import pandas as pd
import os

# Load dataset
df = pd.read_csv("/content/content_refresh_anonymized.csv")

# Fill missing values
df["avg_position"] = df["avg_position"].fillna(0)
df["ctr"] = df["ctr"].fillna(0)
df["search_volume"] = df["search_volume"].fillna(0)
df["content_age_days"] = df["content_age_days"].fillna(0)

# Baseline score
df["baseline_score"] = (
    (df["avg_position"] > 20).astype(int) * 40 +
    (df["ctr"] < 2).astype(int) * 30 +
    (df["search_volume"] > 100).astype(int) * 20 +
    (df["content_age_days"] > 180).astype(int) * 10
)

# Reason code
df["reason_code"] = "REFRESH"

# Action
df["action"] = "CONTENT_REFRESH"

df[["content_id","baseline_score"]].head()
```

## 2. Build the ranked queue (writes the CSV)

*Code the score, rank everything, write work/outputs/baseline_action_score.csv.*

```python
import os

# Sort by score (highest first)
ranked = df.sort_values("baseline_score", ascending=False).copy()

# Keep required columns
ranked["reason_code"] = "REFRESH"
ranked["action"] = "CONTENT_REFRESH"

output = ranked[
    [
        "content_id",
        "client_id",
        "baseline_score",
        "reason_code",
        "action"
    ]
]

# Create output folder
os.makedirs("work/outputs", exist_ok=True)

# Save CSV
output.to_csv(
    "work/outputs/baseline_action_score.csv",
    index=False
)

print("Saved successfully!")
print(output.head(10))
```

## 4. Weak picks + leakage check

*Which picks look wrong and why? Confirm no product flags or future windows leaked in.*

### Weak Picks

Some pages may receive a high baseline score even though they should not be refreshed.

Possible reasons include:

- Temporary ranking fluctuations.
- Seasonal search demand.
- Incorrect or missing CTR values.
- High search volume but already optimized content.

### Leakage Check

The baseline rule only uses historical signals:

- Average Position
- CTR
- Search Volume
- Content Age

It does **not** use future information or label columns such as:

- trend_direction
- trend_pct

Therefore no future-window leakage is introduced.
